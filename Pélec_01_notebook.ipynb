On va analyser les valeurs manquantes, on va afficher un histogramme sur le taux de remplissage.

```python
def tauxRemplissage(df):
    """
    Entrée : DataFrame
    Objectifs : Calcul du taux de remplissage, 1 beaucoup de données, 0 beaucoup de valeurs manquantes
    Sorties : DataFrame
    """
        
    val = 1 - df.isna().sum()/len(df)
    val = pd.DataFrame(val)
    val = val.transpose()
    return val

#######################################

def graphRemplissage(df) :
    """
    Entrée : Dataframe
    Objectifs : Récupérer la DataFrame du taux de remplissage et afficher le graphique du taux de remplissage
    Sortie : Graphe
    """

    df = tauxRemplissage (df)
    fig,ax = plt.subplots()
    ax.tick_params(axis='x', rotation=90)
    sns.barplot( data=df, color = 'blue')
    plt.xlabel('\nColonnes')
    plt.ylabel('Taux de valeurs\n')
    plt.title("Taux de remplissage")
    plt.show()
```

### 2.3 DataFrame unique <a class="anchor" id="section_2_3"></a>

#### 2.3.1 Concaténation  <a class="anchor" id="section_2_3_1"></a>

Avant de concaténer les deux tableaux pour avoir qu’un tableau à analyser, on va supprimer les colonnes qui ne sont pas présentes dans l’autre fichier.<br>
« data_2015 » a un dictionnaire qui contient : « Latitude », « Longitude » et « Human Adress’. Dans « Human Adresse » contient les colonnes : « Address », «City», « State » et « Zip ». Pour être cohérent avec « data_2016 », on va récupérer les valeurs de « Latitude », « Longitude » ; puis les fusionner à 2015 selon « OSEBuildingID ».<br>

```python
data_2015['Location'] = [ast.literal_eval(str(item)) for index, item in data_2015.Location.iteritems()]
data_2015 = pd.concat([data_2015.drop(['Location'], axis=1), data_2015['Location'].apply(pd.Series)], axis=1)
data_2015['human_address'] = [ast.literal_eval(str(item)) for index, item in data_2015.human_address.iteritems()]
data_2015 = pd.concat([data_2015.drop(['human_address'], axis=1), data_2015['human_address'].apply(pd.Series)], axis=1)
data_2015.head()
```

Harmonisation des colonnes et suppression de zip

```python
data_2015.rename(columns={'latitude': 'Latitude','longitude': 'Longitude', 'address': 'Address', 'city': 'City', 'state': 'State'}, inplace=True)
```

Conversion des colonnes "Latitude" et "Longitude" de types "object" en "float".

```python
data_2015['Latitude'] = data_2015['Latitude'].astype(float, errors = 'raise')
data_2015['Longitude'] = data_2015['Longitude'].astype(float, errors = 'raise')
```

Il faut supprimer les colonnes de "data_2015" :<br>
- '2010 Census Tracts',
- 'Seattle Police Department Micro Community Policing Plan Areas',
- 'City Council Districts',
- 'SPD Beats'.

```python
data_2015.drop(['2010 Census Tracts','zip',
'Seattle Police Department Micro Community Policing Plan Areas',
'City Council Districts', 'SPD Beats'], axis = 1, inplace = True)
```

Maintenant que les fichiers ont les mêmes noms de colonnes, les deux tableaux peuvent être concaténés.

```python
df = pd.concat([data_2015,data_2016], join="outer", ignore_index=True, verify_integrity=True)
df.head()
```

#### 2.3.2 Harmonisation des chaînes de caractères <a class="anchor" id="section_2_3_2"></a>

Dans la colonne : « PrimaryPropertyType », on peut voir des valeurs doubles avec une orthographe différente, on va les normaliser. On va mettre la première lettre en majuscule, supprimer le « \n » et gérer les espaces en trop.

```python
df['PrimaryPropertyType'] = df['PrimaryPropertyType'].str.strip("\n")
```

```python
ancien = ["Supermarket / grocery store"]

nouveau = ["Supermarket/grocery Store"]

for i in range(0,len(df['PrimaryPropertyType']),1):
    for j in range(0, len(ancien),1):
        if df['PrimaryPropertyType'][i] == ancien[j]:
            df['PrimaryPropertyType'][i] = nouveau[j]
```

```python
df['Neighborhood'] = df['Neighborhood'].str.capitalize()

for i in range(0,len(df['PrimaryPropertyType']),1):
    if df['Neighborhood'][i] == 'Delridge neighborhoods':
        df['Neighborhood'][i] = 'Delridge'
```

#### 2.3.3 Suppression des valeurs abérantes <a class="anchor" id="section_2_3_3"></a>

On va retirer les lignes qui sont entièrement nulles.

```python
def val_nul(data):
    """
    Entrée : data
    Objectif retirer les lignes nulles
    Sortie : somme sur les lignes
    """
    prov = data.select_dtypes(include = 'float64')
    return prov.sum(axis = 1)
```

```python
val = val_nul(df) != 0
number = np.where(val)
df = df.iloc[number].copy()
```

### 3.1 Distance Harvesine <a class="anchor" id="section_3_1"></a>

On va calculer la distance d’Haversine par rapport au centre de Seattle et afficher le résultat dans un graphique sous forme d’un nuage de point.<br>
Le centre de Seattle est :
- Lattitude : 47.6062095
- Longitude : -122.3320708

```python
centre = [47.6062095,-122.3320708]
```

```python
df_haversine = df.copy()
df_haversine = df_haversine.reset_index()
df_haversine.drop('index',axis = 1, inplace = True)

for i in range(0,len(df_haversine["Longitude"]),1) :
    df_haversine["dist_haversine"] = haversine(centre,(df_haversine["Latitude"][i], df_haversine["Longitude"][i]), unit='mi')
```

### 3.2 Calcul des clusters <a class="anchor" id="section_3_2"></a>

```python
scaler = StandardScaler()
X = np.abs(df[["Latitude", "Longitude"]])
X_scaled = scaler.fit_transform(X)

model = KMeans()
plt.rcParams["figure.figsize"] = (10,5)
visualizer = KElbowVisualizer(model, k =(2, 10))
visualizer.fit(X_scaled)
visualizer.show()
```

Calcul du K-Means et affectation du cluster pour chaque bâtiment selon la « Longitude » et la « Latitude ».

```python
kmeans = KMeans(n_clusters=5)

kmeans.fit(X_scaled)
pred = kmeans.predict(X_scaled)

clusters = pd.DataFrame(pred, index=df["OSEBuildingID"]).reset_index()
clusters.columns = ["OSEBuildingID", "Clusters"]
df = pd.merge(df, clusters, on="OSEBuildingID")
df.drop_duplicates(inplace=True)
df.head()
```

```python
ax = df.plot(kind="scatter", x="Longitude", y="Latitude", figsize=(7,12),
        alpha = 1, c="Clusters", colorbar=False,cmap=plt.get_cmap('Paired'))
plt.xlabel("Longitude", fontsize=14)
plt.ylabel("Latitude", fontsize=14)
plt.title("Localisation des 5 clusters")
plt.show()
```

Avec ce graphique on peut voit les 5 clusters et ainsi avoir un découpage différent.
